```python
#柱状图
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = ['SimHei']
plt.figure(figsize=(10,5))
subject=['math','kow','english']
sorce=[80,90,85]
plt.bar(subject,sorce,label='成绩',width=0.1)
plt.legend(loc='upper left')
plt.title('学生成绩')
plt.ylabel('分数')
plt.xlabel('科目')
plt.grid(axis='y',color='k',linestyle='--',alpha=0.5)
plt.ylim(0,100)
for x,y in zip(subject,sorce):
    plt.text(x,y,y,ha='center',va='bottom',fontsize=20)
plt.show()
```

```python
#饼图的绘制
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = ['SimHei']
plt.figure(figsize=(10,5))
nums=[7,5,6,8]
sub=['语文','数学','政治','物理']
Colors=['red','green','blue','yellow','magenta','cyan']
plt.pie(nums,labels=sub,autopct='%1.1f%%',colors=Colors,wedgeprops={'width':0.5},pctdistance=0.7)
plt.title('科目占比')

plt.tight_layout()
plt.show()
```

```python
#爆炸饼图
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = ['SimHei']
plt.figure(figsize=(10,5))
nums=[7,5,6,8]
sub=['语文','数学','政治','物理']
explode=[0.1,0,0,0]
Colors=['red','green','blue','yellow','magenta','cyan']
plt.pie(nums,labels=sub,autopct='%1.1f%%',colors=Colors,pctdistance=0.7,explode=explode)
plt.title('科目占比')

plt.tight_layout()
plt.show()
```

```python
#散点图
import matplotlib.pyplot as plt
import random
plt.figure(figsize=(10,5))
nums=[]
sub=[]
for i in range(100):
    tem=random.uniform(0,1)
    nums.append(tem)
    sub.append(tem*2+random.gauss(0,1))

plt.scatter(nums,sub,label='<UNK>'
            ,alpha=1,color='red',
            s=50
            )
plt.plot([0,1],[0,1],'k--')
plt.show()
```

```python
#多图表的操作
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
moth=['一月','二月','三月','四月','五月']
val=[100,200,500,300,600]
plt.subplot(2,2,1)
plt.plot(moth,val,label='<UNK>')
plt.subplot(2,2,2)
plt.bar(moth,val,label='<UNK>')
plt.subplot(2,2,3)
plt.barh(moth,val,label='<UNK>')
plt.subplot(2,2,4)
plt.pie(val,labels=moth,autopct='%1.1f%%')
plt.show()
```
