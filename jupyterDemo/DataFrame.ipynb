```python
ds=pd.Series(np.random.randint(10,20,5))
ds.index=pd.Index(['a','b','c','d','e'],name='index')
c=pd.DataFrame({'value':ds,'index':ds.index},index=ds.index,columns=['index','value'])
c
```

```python
#分析学生成绩，1.计算每位学生的总分和平均分，找出数学成绩高于90分或者英语成绩高于85的学生，按照总分从高到低，并输出前三名学生
import pandas as pd
data={
    '姓名':['张三','李四','王五','赵六','钱七'],
    '数学':['85','92','78','88','95'],
    '英语':['90','88','85','92','80'],
    '物理':['75','80','88','85','90']
}

ss=pd.DataFrame(data)
bb=ss[['数学','英语','物理']].astype(int).sum(axis=1)
ss['总分']=bb
ss
```
